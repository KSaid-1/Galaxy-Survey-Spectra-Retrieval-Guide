```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

SDSS
-------
You need "plate,mjd,fiberid" which you can get them through SQL query from casjobs or if you have only have RA and DEC you can use the SDSS crossid here https://skyserver.sdss.org/dr14/en/tools/crossid/crossid.aspx to get them. Once you have that you can either use SAS here https://dr14.sdss.org/optical/spectrum/search to create links that you can save in a file then wget them at once. Alternatively, you can create the links use the below cell

```python
sdss_data = pd.read_csv("example.csv")
sdss_data
```

```python
objid = sdss_data["objid"]; plate = sdss_data["plate"]; mjd = sdss_data["mjd"]; fiberid = sdss_data["fiberid"]
```

```python
for i in range(0,3):
    #print (objid[i],str(fiberid[i]).zfill(4))
    print ("https://data.sdss.org/sas/dr14/sdss/spectro/redux/26/spectra/lite/"+str(plate[i]).zfill(4)+"/spec-"+str(plate[i]).zfill(4)+"-"+str(mjd[i])+"-"+str(fiberid[i]).zfill(4)+".fits")
    os.system("wget -P . -O "+"./spectra/"+str(objid[i])+".fits https://data.sdss.org/sas/dr14/sdss/spectro/redux/26/spectra/lite/"+str(plate[i]).zfill(4)+"/spec-"+str(plate[i]).zfill(4)+"-"+str(mjd[i])+"-"+str(fiberid[i]).zfill(4)+".fits")
```
